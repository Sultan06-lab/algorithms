### Задание №2
Написать генератор случайных матриц(многомерных), который принимает
опциональные параметры <b>m</b>, <b>n</b>, <b>min_limit</b>, <b>max_limit</b>, где <b>m</b> и <b>n</b> указывают размер
матрицы, а <b>min_lim</b> и <b>max_lim</b> - минимальное и максимальное значение для
генерируемого числа.

```python
import random
import time
import math
```

```python
m = int(input());
n = int(input());
min_limit = int(input());
max_limit = int(input());

def matrix(m, n, min_limit, max_limit):
    A = []
    for i in range(m):
        B = []
        for j in range(n):
            B.append(random.randint(min_limit, max_limit))
        A.append(B)
    return A
mat = matrix(m, n, min_limit, max_limit);
for row in mat:
    print(row)
```

### Задание №3
Реализовать методы сортировки строк числовой матрицы в соответствии с
заданием. Оценить время работы каждого алгоритма сортировки и сравнить его со
временем стандартной функции сортировки. Испытания проводить на сгенерированных
матрицах.

```python
import copy
```

```python
# Сортировка выбором.
# Берётся срез массива, в котором минимальный элемент переносят в самый левый угол,
# после чего срез уменьшается и цикл повторяется.
start_time = time.time()
#TODO
def sortSelection(array):
    for i in range(len(array)):
        minIndex = i
        for j in range(i+1, len(array)):
            if array[j] < array[minIndex]:
                minIndex = j
        array[i], array[minIndex] = array[minIndex], array[i]
    return array
for row in mat:
    sortSelection(row)
print("--- {0} ms ---".format(round((time.time() - start_time)*1000)))
```

```python
# Сортировка вставкой.
start_time = time.time()
#TODO
def SortInsertion(array):
    for i in range(len(array)):
        currentItem = array[i]
        position = i
        while position > 0 and currentItem < array[position-1]:
            array[position], array[position-1] = array[position-1], array[position]
            position = position-1
    return array
for row in mat:
    SortInsertion(row)
print("--- {0} ms ---".format(round((time.time() - start_time)*1000)))
```

```python
# Сортировка обменом. || Сортировка пузырьком.
start_time = time.time()
#TODO
def bubbleSort(array):
    for i in range(len(array)):
        for j in range(len(array)-1):
            if(array[j] > array[j+1]):
                array[j], array[j+1] = array[j+1], array[j]
    return array
for row in mat:
    bubbleSort(row)
print("--- {0} ms ---".format(round((time.time() - start_time)*1000)))
```

```python
# Сортировка Шелла.
start_time = time.time()
#TODO
def shellSort(array):
    l=len(array)
    gap = l//2
    while gap > 0: 
        for i in range(gap, l):
            position = i
            while position >= gap and array[position] < array[position-gap]:
                array[position], array[position-gap] = array[position-gap], array[position]
                position = position-gap
        gap = gap//2
    return array;
for row in mat:
    shellSort(row)

print("--- {0} ms ---".format(round((time.time() - start_time)*1000)))
```

```python
# Быстрая сортировка.
start_time = time.time()
#TODO
def quickSort(array):
    l = len(array)
    if len(array)<1:
        return array
    middle = array[math.floor(l/2)];
    startArray = [];
    middleArray = [];
    endArray = [];
    for i in array:
        if i == middle:
            middleArray.append(i)
        if i < middle:
            startArray.append(i)
        if i > middle:
            endArray.append(i)
    return quickSort(startArray) + middleArray  + quickSort(endArray);   
for row in mat:
    quickSort(row)
print("--- {0} ms ---".format(round((time.time() - start_time)*1000)))
    
```

```python
# Турнирная сортировка.
start_time = time.time()
#TODO
def heapify(mas,i,lenn):#меняет родительский узел с дочерним, если он больше
    i_left = 2*i+1
    i_right= 2*i+2
    i_largest = i
    if i_left<=lenn and mas[i_left]>mas[i_largest]:
        i_largest=i_left
    if i_right<=lenn and mas[i_right]>mas[i_largest]:
        i_largest=i_right
    if i_largest == i:
        return
    else: 
        mas[i_largest], mas[i] = mas[i], mas[i_largest]
        heapify(mas,i_largest,lenn)#рекурсия для нового узла
def buildHeap(mas):#строим саму пирамиду
    i_mid = len(mas)//2
    for idx in reversed(range(0,i_mid+1)):
        heapify(mas,idx,len(mas)-1)
def heapSort(mas):#меняем элемент и перестраиваем пирамиду
    buildHeap(mas)
    for idx in reversed(range(0,len(mas))):
        mas[0], mas[idx] = mas[idx], mas[0]
        heapify(mas,0,idx-1)
    return mas
for row in mat:
    heapSort(row)
print("--- {0} ms ---".format(round((time.time() - start_time)*1000)))
```
